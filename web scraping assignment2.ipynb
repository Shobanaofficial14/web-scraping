```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.flipkart.com/mobiles/apple~brand/pr?sid=tyy,4io"
```

```python
permit=requests.get(url)
```

```python
permit
```

```python
data=BeautifulSoup(permit.text)
```

```python
data
```

```python
mobile_Brand=data.select("._4rR01T")
```

```python
mobile_Brand
```

```python
mobile_rating=data.select("._2_R_DZ span span:nth-child(1)")
```

```python
mobile_rating
```

```python
mobile_Price=data.select("._1_WHN1")
```

```python
mobile_Price
```
